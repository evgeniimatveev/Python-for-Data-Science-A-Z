```python
# ðŸ“š **Lesson: The "If" Statement in Python**
# ----------------------------------------

# ðŸš€ **Introduction**
# The `if` statement is one of the most fundamental concepts in programming. 
# It allows you to control the flow of your program by executing different code blocks based on conditions.

# Structure:
# if <condition>:
#     <code to execute if condition is True>
# elif <another condition>:
#     <code to execute if the second condition is True>
# else:
#     <code to execute if none of the conditions are True>

# ----------------------------------------
# **Section 1: Basics of the "If" Statement** ðŸŒŸ
# ----------------------------------------

# Example 1: Simple `if` statement
x = 10
if x > 5:
    print("x is greater than 5")

# Example 2: `if-else` statement
y = -3
if y >= 0:
    print("y is non-negative")
else:
    print("y is negative")

# âœ… **Practice Task**:
# - Modify the first example to check if `x` is less than or equal to 5.
# - Write an `if-else` block that checks if a number is even or odd.

# ----------------------------------------
# **Section 2: Using `elif` for Multiple Conditions** ðŸ”„
# ----------------------------------------

# Example: Checking number range
z = 0
if z > 0:
    print("z is positive")
elif z == 0:
    print("z is zero")
else:
    print("z is negative")

# âœ… **Practice Task**:
# - Extend the example to include a new condition: check if `z` is greater than 100.
# - Write a program that categorizes a person's age (child, teenager, adult, senior).

# ----------------------------------------
# **Section 3: Nested "If" Statements** ðŸ› ï¸
# ----------------------------------------

# Example: Nested conditions
score = 85
if score >= 50:
    if score >= 90:
        print("Excellent")
    elif score >= 70:
        print("Good")
    else:
        print("Pass")
else:
    print("Fail")

# âœ… **Practice Task**:
# - Write a nested `if` statement to check:
#   - If a number is divisible by 2.
#   - If itâ€™s also divisible by 3.

# ----------------------------------------
# **Section 4: Logical Operators in Conditions** ðŸ”‘
# ----------------------------------------

# Logical operators (`and`, `or`, `not`) can be combined with `if` statements.

# Example: Combining conditions
age = 25
income = 3000
if age > 18 and income >= 2000:
    print("Eligible for a loan")

# Example: Negating conditions
is_sunny = False
if not is_sunny:
    print("Take an umbrella")

# âœ… **Practice Task**:
# - Write an `if` statement to check if a number is between 10 and 20 (inclusive).
# - Use the `not` operator to check if a string is empty.

# ----------------------------------------
# **Section 5: Mini Project: Random Number Categorizer** ðŸŽ¯
# ----------------------------------------

# Letâ€™s create a program that generates a random number and categorizes it.
print("\nMini Project: Random Number Categorizer")

import random

# Generate a random number between -10 and 10
random_number = random.randint(-10, 10)
print("Random Number:", random_number)

# Categorize the number
if random_number > 5:
    print("The number is greater than 5")
elif -5 <= random_number <= 5:
    print("The number is between -5 and 5")
else:
    print("The number is less than -5")

# âœ… **Practice Task**:
# - Extend the project to categorize numbers into more specific ranges (e.g., `> 8`, `-8 to -5`, etc.).
# - Add a loop to generate multiple random numbers and categorize each.

# ----------------------------------------
# **Section 6: Advanced Example: Temperature Converter** ðŸŒ¡ï¸
# ----------------------------------------

# Letâ€™s write a program to categorize temperatures.
print("\nAdvanced Example: Temperature Converter")

temperature = float(input("Enter the temperature in Celsius: "))

if temperature > 30:
    print("It's hot")
elif 15 <= temperature <= 30:
    print("It's warm")
elif 0 <= temperature < 15:
    print("It's cold")
else:
    print("It's freezing!")

# âœ… **Practice Task**:
# - Extend the program to also convert the temperature to Fahrenheit.
# - Add a condition for extreme heat (e.g., temperature > 40).
```
