```python
#  Advanced Data Visualization with Matplotlib
# ============================================
# This script explores advanced techniques in Matplotlib, including:
# - Subplots for comparing multiple datasets.
# - Customizing plot aesthetics for polished results.
# - Adding annotations and interactive elements.
# Let's get started! 

# Importing required libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn
# ------------------------------
#   Preparing Data
# ------------------------------
# Sample data: Seasons and player stats
Seasons = np.arange(2005, 2015)
Kobe_Games = [80, 77, 82, 73, 82, 58, 78, 6, 35, 50]
LeBron_Games = [79, 78, 75, 81, 76, 79, 62, 76, 77, 69]
Durant_Games = [35, 80, 74, 82, 82, 78, 66, 81, 81, 27]

# Sample points scored
Kobe_Points = [2832, 2430, 2323, 2201, 1970, 1616, 1171, 83, 666, 782]
LeBron_Points = [2478, 2132, 2250, 2304, 2089, 2254, 1683, 2036, 2089, 1743]
Durant_Points = [903, 1806, 1951, 2322, 2479, 1930, 2280, 2593, 2592, 686]

Players = ["Kobe Bryant", "LeBron James", "Kevin Durant"]

# ------------------------------
#  Subplots for Comparisons
# ------------------------------
# Create subplots for games played and points scored
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# First subplot: Games played
axs[0].plot(Seasons, Kobe_Games, label="Kobe Bryant", color="purple", marker="o")
axs[0].plot(Seasons, LeBron_Games, label="LeBron James", color="orange", marker="s")
axs[0].plot(Seasons, Durant_Games, label="Kevin Durant", color="green", marker="^")
axs[0].set_title("Games Played Over Seasons", fontsize=16)
axs[0].set_ylabel("Games Played", fontsize=14)
axs[0].legend(loc="upper left")
axs[0].grid(True, linestyle="--", linewidth=0.5)

# Second subplot: Points scored
axs[1].plot(Seasons, Kobe_Points, label="Kobe Bryant", color="purple", marker="o")
axs[1].plot(Seasons, LeBron_Points, label="LeBron James", color="orange", marker="s")
axs[1].plot(Seasons, Durant_Points, label="Kevin Durant", color="green", marker="^")
axs[1].set_title("Points Scored Over Seasons", fontsize=16)
axs[1].set_xlabel("Season", fontsize=14)
axs[1].set_ylabel("Points Scored", fontsize=14)
axs[1].legend(loc="upper left")
axs[1].grid(True, linestyle="--", linewidth=0.5)

# Adjust layout
plt.tight_layout()
plt.show()

# ------------------------------
#   Highlight Key Data Points with Annotations
# ------------------------------
# Create a single plot with annotations
plt.figure(figsize=(12, 6))

# Plot data for Kobe Bryant
plt.plot(Seasons, Kobe_Points, label="Kobe Bryant", color="purple", marker="o")

# Add annotation for key event
plt.annotate("Injury Season", xy=(2013, 666), xytext=(2010, 1500),
             arrowprops=dict(facecolor="red", arrowstyle="->"), fontsize=12)

# Add chart details
plt.title("Kobe Bryant: Points Scored with Key Highlight", fontsize=16)
plt.xlabel("Season", fontsize=14)
plt.ylabel("Points Scored", fontsize=14)
plt.legend(loc="upper left", fontsize=12)
plt.grid(True, linestyle="--", linewidth=0.5)

# Show the plot
plt.show()



# ------------------------------
#  Mini Project: Efficiency Analysis
# ------------------------------
# Calculate Points Per Game
Kobe_PPG = np.array(Kobe_Points) / np.array(Kobe_Games)
LeBron_PPG = np.array(LeBron_Points) / np.array(LeBron_Games)
Durant_PPG = np.array(Durant_Points) / np.array(Durant_Games)

# Plot Points Per Game
plt.figure(figsize=(12, 6))
plt.plot(Seasons, Kobe_PPG, label="Kobe Bryant", color="purple", marker="o", linestyle="--")
plt.plot(Seasons, LeBron_PPG, label="LeBron James", color="orange", marker="s", linestyle="-")
plt.plot(Seasons, Durant_PPG, label="Kevin Durant", color="green", marker="^", linestyle=":")

# Add labels and title
plt.title("Points Per Game Over Seasons", fontsize=16)
plt.xlabel("Season", fontsize=14)
plt.ylabel("Points Per Game", fontsize=14)
plt.legend(loc="upper left", fontsize=12)

# Add a grid and show the plot
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.show()
```
