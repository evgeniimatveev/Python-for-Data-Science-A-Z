```python
#  Histograms in Seaborn: Visualizing Distributions 

# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

# ------------------------------
# 1 Loading and Exploring the Data
# ------------------------------
# Load the dataset
movies = pd.read_csv("Movie-Ratings(S5).csv")

# Rename columns for better readability
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']

# Display basic information about the dataset
print("Dataset Overview:")
print(movies.info())
print("\nFirst 5 Rows:")
print(movies.head())

# ------------------------------
# 2 Simple Histogram: Using Seaborn
# ------------------------------
# Create a histogram for 'CriticRating' with 15 bins
sns.displot(
    movies['CriticRating'],  # Data column
    bins=15,                 # Number of bins
    kde=True,                # Add Kernel Density Estimation curve
    color="blue",            # Customize the color
    height=6                 # Adjust the figure size
)
plt.title("Distribution of Critic Ratings ")
plt.xlabel("Critic Ratings")
plt.ylabel("Frequency")
plt.show()

# ------------------------------
# 3 Simple Histogram: Using Matplotlib
# ------------------------------
# Create a histogram for 'AudienceRating' with Matplotlib
plt.hist(
    movies['AudienceRating'],  # Data column
    bins=15,                   # Number of bins
    color="skyblue",           # Customize the color
    edgecolor="black"          # Add border to the bins
)
plt.title("Distribution of Audience Ratings ")
plt.xlabel("Audience Ratings")
plt.ylabel("Frequency")
plt.show()

# ------------------------------
# 4 Advanced Histogram: Styling with Seaborn
# ------------------------------
# Customizing bins, colors, and adding KDE for smoothness
sns.histplot(
    data=movies,
    x='BudgetMillions',       # Column for histogram
    bins=20,                  # Number of bins
    kde=True,                 # Add a KDE curve
    color="green",            # Customize the color
    line_kws={"linewidth": 2} # Line styling for KDE
)
plt.title("Distribution of Movie Budgets ")
plt.xlabel("Budget (Millions)")
plt.ylabel("Frequency")
plt.show()

# ------------------------------
# 5 Layering Histograms for Comparison
# ------------------------------
# Compare 'CriticRating' and 'AudienceRating' on the same plot
sns.histplot(
    data=movies,
    x='CriticRating',
    bins=15,
    color="blue",
    kde=True,
    label="Critic Rating"     # Add a legend label
)
sns.histplot(
    data=movies,
    x='AudienceRating',
    bins=15,
    color="orange",
    kde=True,
    label="Audience Rating"   # Add a legend label
)
plt.title("Comparison of Critic and Audience Ratings ")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.legend()  # Display the legend
plt.show()

# ------------------------------
#  Summary and Insights
# ------------------------------
"""
1 **Basic Histograms**:
   - Use `sns.displot` or `plt.hist` for simple distribution analysis.
   - Adjust the number of bins (`bins=15`) to refine data granularity.

2 **KDE (Kernel Density Estimation)**:
   - Adding `kde=True` provides a smooth curve to understand data patterns.

3 **Advanced Customization**:
   - Use `sns.histplot` for more styling options, such as line thickness and colors.

4 **Layered Histograms**:
   - Compare multiple distributions (e.g., Critic vs Audience Ratings) on the same chart.

 **Practice Idea**:
   - Group the data by `Genre` and create separate histograms for each category.
   - Experiment with other columns like `Year` or `BudgetMillions`.
"""
```
