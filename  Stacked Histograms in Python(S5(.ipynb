```python
# 🟦 Stacked Histograms in Python: Exploring Multiple Categories Simultaneously 🎯

# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

# ------------------------------
# 1️⃣ Loading the Dataset
# ------------------------------
# Load the dataset and adjust the column names
movies = pd.read_csv("Movie-Ratings(S5).csv")
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']

# Display dataset summary
print("Dataset Information:")
print(movies.info())
print("\nFirst 5 Rows of Data:")
print(movies.head())

# ------------------------------
# 2️⃣ Individual Histograms by Genre
# ------------------------------
# Plotting individual histograms for specific genres
plt.hist(
    movies[movies.Genre == 'Action'].BudgetMillions, bins=15, color="blue", label="Action", alpha=0.7
)
plt.hist(
    movies[movies.Genre == 'Drama'].BudgetMillions, bins=15, color="green", label="Drama", alpha=0.7
)
plt.hist(
    movies[movies.Genre == 'Thriller'].BudgetMillions, bins=15, color="red", label="Thriller", alpha=0.7
)
plt.title("Individual Histograms for Selected Genres 🎥")
plt.xlabel("Budget (Millions)")
plt.ylabel("Frequency")
plt.legend()  # Adding legend
plt.show()

# ------------------------------
# 3️⃣ Stacked Histogram: Matplotlib
# ------------------------------
# Creating a stacked histogram with multiple genres
plt.hist(
    [
        movies[movies.Genre == 'Action'].BudgetMillions,
        movies[movies.Genre == 'Drama'].BudgetMillions,
        movies[movies.Genre == 'Comedy'].BudgetMillions,
    ],
    bins=15,
    stacked=True,           # Enable stacking
    color=["blue", "green", "orange"],  # Colors for each genre
    label=["Action", "Drama", "Comedy"] # Labels for the legend
)
plt.title("Stacked Histogram for Multiple Genres 📊")
plt.xlabel("Budget (Millions)")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# ------------------------------
# 4️⃣ Automating Stacked Histograms for All Genres
# ------------------------------
# Generate a stacked histogram for all genres programmatically
list1 = []   # To store budget data for each genre
mylabels = []  # To store labels (genres)

# Loop through each unique genre and collect data
for gen in movies.Genre.unique():
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)  # Add genre to labels

# Plot the stacked histogram
plt.hist(
    list1,                # List of data for stacking
    bins=30,              # Number of bins
    stacked=True,         # Enable stacking
    rwidth=1,             # Adjust bar width
    label=mylabels        # Apply labels
)
plt.title("Stacked Histogram for All Genres 🎭")
plt.xlabel("Budget (Millions)")
plt.ylabel("Frequency")
plt.legend(loc="upper right")  # Position legend
plt.show()

# ------------------------------
# 📚 Summary and Insights
# ------------------------------
"""
1️⃣ **Stacked Histograms**:
   - Allows comparison of multiple distributions (e.g., movie budgets across genres).
   - Use `stacked=True` in Matplotlib to stack multiple datasets.

2️⃣ **Automation**:
   - Automate the process with loops to include all categories dynamically (e.g., all movie genres).

3️⃣ **Customizations**:
   - Adjust transparency (`alpha`), bar width (`rwidth`), and colors for better clarity.

4️⃣ **Key Use Case**:
   - Compare subsets of data (e.g., budget distributions for Action vs Drama movies).

🎯 **Advanced Idea**:
   - Combine stacked histograms with KDE plots to overlay smoothed distributions.
"""
```
