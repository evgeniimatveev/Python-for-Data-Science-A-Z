```python
#  Introduction to Seaborn: A Python Data Visualization Library 
# Seaborn is a powerful library for creating informative and attractive statistical graphics.
# It builds on top of Matplotlib and integrates well with Pandas data structures.

# -----------------------------
# 1 Importing Seaborn and Other Libraries
# -----------------------------
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Enable inline plotting (for Jupyter Notebook)
%matplotlib inline

# -----------------------------
# 2 Creating a Sample Dataset
# -----------------------------
# Seaborn works seamlessly with Pandas DataFrames. Let's create one for demonstration.
data = {
    "Country": ["USA", "India", "China", "Germany", "Brazil"],
    "GDP": [21.43, 2.87, 14.34, 3.85, 2.05],  # Trillions
    "LifeExpectancy": [78.8, 69.7, 76.7, 81.1, 75.7],  # Years
    "InternetUsers": [89, 55, 65, 92, 70],  # Percentage
}

df = pd.DataFrame(data)

# Display the DataFrame
print("Sample Dataset:")
print(df)

# -----------------------------
# 3 Creating a Simple Distribution Plot
# -----------------------------
# Distribution plots show the distribution of a single variable.
plt.figure(figsize=(8, 6))
sns.histplot(df["LifeExpectancy"], kde=True, color="skyblue", bins=10)
plt.title("Distribution of Life Expectancy", fontsize=16)
plt.xlabel("Life Expectancy (Years)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# -----------------------------
# 4 Visualizing Relationships with Scatter Plots
# -----------------------------
# Scatter plots show relationships between two variables.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="GDP", y="LifeExpectancy", hue="Country", palette="viridis", s=100)
plt.title("GDP vs. Life Expectancy", fontsize=16)
plt.xlabel("GDP (Trillions)", fontsize=12)
plt.ylabel("Life Expectancy (Years)", fontsize=12)
plt.legend(title="Country", fontsize=10)
plt.show()

# -----------------------------
# 5 Box Plot: Distribution Across Categories
# -----------------------------
# Box plots show distributions across categorical data.
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Country", y="InternetUsers", palette="pastel")
plt.title("Internet Users by Country", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Internet Users (%)", fontsize=12)
plt.show()

# -----------------------------
# 6 Pairplot: Multi-variable Relationships
# -----------------------------
# Pairplots visualize relationships for all pairwise combinations of numerical columns.
sns.pairplot(df, diag_kind="kde", kind="scatter", hue="Country", palette="coolwarm")
plt.suptitle("Pairplot of Variables", y=1.02, fontsize=16)
plt.show()

# -----------------------------
#  Why Use Seaborn? 
# -----------------------------
"""
1 **Ease of Use**: Seaborn simplifies complex plots like violin plots or heatmaps with just one line of code.
2 **Beautiful Defaults**: Attractive color palettes and styles reduce the need for customization.
3 **Statistical Features**: Built-in functions to plot regression lines, visualize distributions, and much more.
4 **Integration**: Works seamlessly with Pandas and Numpy, making it ideal for data analysis workflows.
"""

# -----------------------------
#  Key Takeaway
# -----------------------------
"""
Seaborn is a fantastic library for creating professional and informative visualizations in Python.
Its simplicity and functionality make it perfect for both beginners and advanced users.
"""

#  Bonus: Explore the Seaborn Gallery!
# Visit https://seaborn.pydata.org/examples/index.html to see amazing examples and get inspired.
```
