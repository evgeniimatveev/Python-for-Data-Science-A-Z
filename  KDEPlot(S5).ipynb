```python
# Importing required libraries 📚
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Creating a sample dataset 🎬
# Dataset includes genres, critic ratings, and audience ratings.
movies = pd.DataFrame({
    'Genre': ['Action', 'Comedy', 'Drama', 'Horror', 'Thriller', 'Romance'] * 10,
    'CriticRating': [50, 60, 70, 80, 90, 45] * 10,
    'AudienceRating': [30, 75, 50, 60, 85, 55] * 10
})

# Converting 'Genre' to a categorical data type for optimization 🏷️
# This helps save memory and improves performance.
movies['Genre'] = movies['Genre'].astype('category')

# Step 2: KDE Plot (Kernel Density Estimation) for a single variable 🎥
# Plot: Displays the distribution of CriticRating using a KDE plot.
# Customization: Adds shading (shade=True) and sets the color to blue.
# Insight: Identifies areas where critic ratings are concentrated.
sns.kdeplot(data=movies, x='CriticRating', shade=True, color='blue')
plt.title("KDE Plot: Critic Rating Distribution")  # Title of the plot 🖼️
plt.show()

# Step 3: 2D KDE Plot for two variables 🎞️
# Plot: Shows the density of two variables, CriticRating and AudienceRating.
# Customization:
#   - Colormap (cmap): Enhances visuals using the coolwarm palette.
#   - Color bar (cbar=True): Indicates density levels.
# Insight: Highlights clusters in the ratings to reveal relationships.
sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    cmap='coolwarm',
    shade=True,
    cbar=True
)
plt.title("2D KDE Plot: Critic vs Audience Ratings")  # Adding a title 🖌️
plt.show()

# Step 4: Scatter Plot with genres highlighted 📊
# Plot: Scatter plot showing CriticRating vs AudienceRating for all movies.
# Customization:
#   - hue and style: Differentiate genres using colors and marker styles.
#   - palette='Set2': Aesthetic color scheme for better visibility.
#   - Larger points (s=100) for clarity.
# Insight: Easy genre-wise comparison of audience and critic ratings.
sns.scatterplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    hue='Genre',
    style='Genre',
    palette='Set2',
    s=100  # Increasing marker size for better visibility
)
plt.title("Scatter Plot: Critic vs Audience Ratings by Genre")
plt.show()

# Step 5: Advanced KDE Plot with heatmap effect 🔥
# Plot: Combines density estimation and heatmap in one chart.
# Customization:
#   - Color Map (cmap): Red gradient for heatmap.
#   - Shading (shade=True): Adds filled contours.
#   - shade_lowest=False: Hides the least dense areas.
#   - Adds transparency (alpha=0.7).
# Insight: Visualizes both density and intensity of ratings for deeper analysis.
sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    shade=True,
    shade_lowest=False,
    cmap='Reds',
    alpha=0.7
)
plt.title("Customized KDE Plot: Critic vs Audience Ratings")  # Title added
plt.show()
```
