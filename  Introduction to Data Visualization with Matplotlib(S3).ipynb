```python
#  Introduction to Data Visualization with Matplotlib
# =======================================================
# Welcome to your first visualization lecture using Matplotlib! 
# Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python. 

# ------------------------------
#  1. Importing Libraries
# ------------------------------

# Importing NumPy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Ensure inline plotting for Jupyter Notebooks
%matplotlib inline

# ------------------------------
#  2. Preparing Data
# ------------------------------

# Create some sample data to visualize
Players = ["Kobe Bryant", "LeBron James", "Kevin Durant"]
Seasons = list(range(2005, 2015))
Salaries = np.array([
    [15946875, 17718750, 19490625, 21262500, 23034375, 24806250, 25244493, 27849149, 30453805, 23500000],
    [4621800, 5186160, 5567160, 6054000, 7725000, 13740400, 15384950, 17545000, 19300000, 20707000],
    [0, 0, 0, 4156674, 4302022, 1739842, 15000000, 16441000, 18771000, 20141000]
])

# ------------------------------
#  3. First Visualization
# ------------------------------

# Setting up the figure size for better visibility
plt.rcParams['figure.figsize'] = (10, 6)

# Plotting Salary Data for Kobe Bryant
plt.plot(Salaries[0], color='Black', linestyle='--', marker='s', markersize=7, label=Players[0])

# Adding labels and title
plt.title("Salaries Over Seasons", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Salary ($)", fontsize=12)
plt.xticks(ticks=range(len(Seasons)), labels=Seasons, rotation=45)

# Adding legend
plt.legend(loc='upper left')

# Display the plot
plt.show()

# ------------------------------
#  4. Adding More Players to the Visualization
# ------------------------------

# Plotting Salary Data for All Players
for i in range(len(Players)):
    plt.plot(Salaries[i], linestyle='--', marker='o', markersize=7, label=Players[i])

# Adding labels and title
plt.title("Salaries Over Seasons for All Players", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Salary ($)", fontsize=12)
plt.xticks(ticks=range(len(Seasons)), labels=Seasons, rotation=45)

# Adding legend
plt.legend(loc='upper left')

# Display the plot
plt.show()

# ------------------------------
#  5. Visualization Insights
# ------------------------------
# From the graphs above, we can see trends in player salaries over time. 
#  Kobe Bryant's salary peaked in 2014.
#  LeBron James shows consistent salary growth.
#  Kevin Durant had a significant increase starting from 2013.

# ------------------------------
#  6. Customizing Plots
# ------------------------------

# Let's make a customized plot for better aesthetics
plt.figure(figsize=(12, 8))

# Plot all players
colors = ['Black', 'Blue', 'Green']
styles = ['-', '--', ':']
for i, player in enumerate(Players):
    plt.plot(Salaries[i], color=colors[i], linestyle=styles[i], marker='o', label=player)

# Add grid for readability
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Adding labels, title, and legend
plt.title("Player Salaries Over Time (Customized)", fontsize=16)
plt.xlabel("Season", fontsize=14)
plt.ylabel("Salary ($)", fontsize=14)
plt.xticks(ticks=range(len(Seasons)), labels=Seasons, rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='upper left', fontsize=12)

# Show the plot
plt.show()

# ------------------------------
#  Mini Project: Compare Performance with Salary
# ------------------------------

# Example: Calculate Field Goals Per Salary ($ per 100k)
FieldGoals = np.array([
    [978, 813, 716, 800, 694, 574, 345, 295, 283, 50],
    [875, 913, 798, 806, 798, 725, 767, 624, 625, 603],
    [306, 240, 161, 319, 509, 712, 720, 731, 754, 849]
])

FieldGoalsPerSalary = FieldGoals / (Salaries / 100000)

# Visualize the performance metric
plt.figure(figsize=(12, 8))
for i, player in enumerate(Players):
    plt.plot(FieldGoalsPerSalary[i], color=colors[i], linestyle=styles[i], marker='o', label=player)

plt.title("Field Goals per $100k Salary", fontsize=16)
plt.xlabel("Season", fontsize=14)
plt.ylabel("Field Goals / $100k", fontsize=14)
plt.xticks(ticks=range(len(Seasons)), labels=Seasons, rotation=45, fontsize=12)
plt.legend(loc='upper right', fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()

# ------------------------------
#  Conclusion
# ------------------------------
# In this lecture, we learned:
# - How to create basic visualizations using Matplotlib
# - Customizing plots for better clarity
# - A real-world use case: player performance vs salary.
# Keep practicing and experimenting to master data visualization! 
```
