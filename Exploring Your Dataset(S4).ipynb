```python
# üìù Lecture Script: Exploring Your Dataset
# ==========================================
# In this script:
# - Use `len()` to explore dataset dimensions.
# - Learn about dataset structure and columns.
# - Add advanced examples for better insights.
# Let‚Äôs explore! üöÄ

# ------------------------------
# üîπ 1. Import Required Libraries
# ------------------------------
import pandas as pd
import os

# Load the dataset (replace 'DemographicData.csv' with your file if needed)
file_path = 'DemographicData.csv'

if os.path.exists(file_path):
    stats = pd.read_csv(file_path)
    print("‚úÖ Dataset successfully loaded!")
else:
    print("‚ö†Ô∏è File not found. Please check the path!")

# ------------------------------
# üîπ 2. Check the Number of Rows and Columns
# ------------------------------
# Total number of rows
print("\nüîç Total Rows:", len(stats))  

# Total number of columns
print("\nüîç Total Columns:", len(stats.columns))  

# ------------------------------
# üîπ 3. Explore Dataset Columns
# ------------------------------
# Display all column names
print("\nüîç Columns in the dataset:")
print(stats.columns)

# Check data types of each column
print("\nüîç Data Types:")
print(stats.dtypes)

# ------------------------------
# üîπ 4. Perform Basic Data Exploration
# ------------------------------
# View a summary of the dataset
print("\nüîç Dataset Overview:")
print(stats.info())

# Summary statistics
print("\nüîç Summary Statistics:")
print(stats.describe())

# ------------------------------
# üîπ 5. Advanced Analysis: Unique Values and Value Counts
# ------------------------------
# Count unique values in the "Income Group" column
print("\nüîç Unique Income Groups:")
print(stats['Income Group'].unique())

# Count occurrences of each "Income Group"
print("\nüîç Value Counts for Income Groups:")
print(stats['Income Group'].value_counts())

# ------------------------------
# üîπ 6. Use `len()` for Specific Filters
# ------------------------------
# Filter countries with Birth Rate > 30 and count them
high_birth_rate = stats[stats['Birth rate'] > 30]
print("\nüîç Number of Countries with Birth Rate > 30:", len(high_birth_rate))

# ------------------------------
# üîπ 7. Display Filtered Data
# ------------------------------
# Show first few rows of filtered data
print("\nüîç High Birth Rate Countries:")
print(high_birth_rate[['Country Name', 'Birth rate']].head())

# ------------------------------
# üîπ 8. Bonus: Identify Empty (NaN) Values
# ------------------------------
# Count total NaN values per column
print("\nüîç Missing Values in Each Column:")
print(stats.isna().sum())

# ------------------------------
# üîπ 9. Visualization: Row and Column Dimensions
# ------------------------------
import matplotlib.pyplot as plt

# Bar chart for dataset dimensions
plt.figure(figsize=(8, 5))
plt.bar(['Rows', 'Columns'], [len(stats), len(stats.columns)], color=['skyblue', 'orange'])
plt.title('Dataset Dimensions')
plt.ylabel('Count')
plt.show()
```
