```python
# üìù Lecture Script: Importing Data into Python
# ============================================
# In this script:
# - Import data using Pandas.
# - Manage working directories.
# - Perform basic data analysis.
# Let‚Äôs get started! üöÄ

# ------------------------------
# üîπ 1. Import Required Libraries
# ------------------------------
import pandas as pd
import os
import matplotlib.pyplot as plt

# ------------------------------
# üîπ 2. Check and Set the Working Directory
# ------------------------------
# Check the current working directory
print("Current Working Directory:", os.getcwd())

# Specify the file path (replace with your file path if needed)
file_path = 'DemographicData.csv'

# Check if the file exists in the directory
if not os.path.exists(file_path):
    print(f"‚ö†Ô∏è File '{file_path}' not found! Ensure the file path is correct.")
else:
    print(f"‚úÖ File '{file_path}' found.")

# ------------------------------
# üîπ 3. Import Data
# ------------------------------
# Load the data
try:
    data = pd.read_csv(file_path)
    print("‚úÖ Data successfully loaded.")
except Exception as e:
    print("‚ö†Ô∏è Error loading the file:", e)

# ------------------------------
# üîπ 4. Initial Data Analysis
# ------------------------------
# Display the first 5 rows
print("\nüîç First 5 rows:")
print(data.head())

# Display dataset information
print("\nüîç Dataset Info:")
print(data.info())

# Display statistical summary
print("\nüîç Statistical Summary:")
print(data.describe())

# ------------------------------
# üîπ 5. Working with Columns and Filtering
# ------------------------------
# Access two specific columns
print("\nüîç Selecting Columns (Country Name and Internet Users):")
print(data[['Country Name', 'Internet users']].head())

# Filter countries with Internet users > 70%
filtered_data = data[data['Internet users'] > 70]
print("\nüîç Countries with Internet Users > 70%:")
print(filtered_data)

# ------------------------------
# üîπ 6. Data Visualization
# ------------------------------
# Plot the relationship between Internet Users and Birth Rate
plt.figure(figsize=(10, 6))
plt.scatter(data['Internet users'], data['Birth rate'], color='purple', alpha=0.6)

# Add labels and title
plt.title("Internet Users vs Birth Rate", fontsize=16)
plt.xlabel("Internet Users (%)", fontsize=12)
plt.ylabel("Birth Rate", fontsize=12)
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```
