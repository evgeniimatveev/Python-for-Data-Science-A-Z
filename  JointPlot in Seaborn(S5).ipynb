```python
# 📊 JointPlot in Seaborn: Exploring Relationships and Distributions 🎯

# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

# ------------------------------
# 1️⃣ Loading and Preprocessing Data
# ------------------------------
# Load the dataset
movies = pd.read_csv("Movie-Ratings(S5).csv")

# Rename columns for readability
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']

# Display basic information about the dataset
print("Dataset Overview:")
print(movies.info())
print("\nFirst 5 Rows:")
print(movies.head())

# ------------------------------
# 2️⃣ Basic JointPlot: Scatter Plot
# ------------------------------
# Create a scatter plot with marginal histograms
sns.jointplot(
    data=movies, 
    x='CriticRating', 
    y='AudienceRating', 
    kind='scatter',  # Scatter plot type
    height=7         # Plot size
)
plt.show()

# ------------------------------
# 3️⃣ Hex Plot: Visualizing Density
# ------------------------------
# Create a hexbin plot to analyze data density
sns.jointplot(
    data=movies, 
    x='CriticRating', 
    y='AudienceRating', 
    kind='hex',       # Hexbin plot type
    height=7, 
    color="purple"    # Customize color
)
plt.show()

# ------------------------------
# 4️⃣ KDE Plot: Smooth Density
# ------------------------------
# Use KDE (Kernel Density Estimation) for smoother distributions
sns.jointplot(
    data=movies, 
    x='CriticRating', 
    y='AudienceRating', 
    kind='kde',       # KDE plot type
    height=7, 
    fill=True         # Fill the KDE plot
)
plt.show()

# ------------------------------
# 5️⃣ Regression Plot: Trends and Relationships
# ------------------------------
# Add a regression line to analyze trends
sns.jointplot(
    data=movies, 
    x='CriticRating', 
    y='AudienceRating', 
    kind='reg',       # Regression plot type
    height=7, 
    color="green"     # Regression line color
)
plt.show()

# ------------------------------
# 6️⃣ Advanced Customization: Marginal Histograms
# ------------------------------
# Customize the marginal histograms and scatter plot
j = sns.jointplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    kind='scatter',   # Scatter plot
    marginal_kws=dict(bins=20, fill=True, color="skyblue"),  # Marginal histograms
    height=7
)
j.fig.suptitle("Critic Rating vs Audience Rating", y=1.03)  # Add a title to the plot
plt.show()

# ------------------------------
# 📚 Summary and Insights
# ------------------------------
"""
1️⃣ **Scatter Plot**: Use this to visualize individual data points and their marginal distributions.
2️⃣ **Hex Plot**: Ideal for large datasets where points overlap, as it highlights density regions.
3️⃣ **KDE Plot**: Great for visualizing smooth distributions and patterns in the data.
4️⃣ **Regression Plot**: Helps identify trends and linear relationships between variables.
5️⃣ **Customizations**: Use marginal_kws to refine the plot aesthetics.

🎯 **Practice Idea**:
- Group the data by "Genre" and use `hue` to analyze relationships for each category.
- Try replacing "CriticRating" or "AudienceRating" with another column, like "BudgetMillions".
"""
```
